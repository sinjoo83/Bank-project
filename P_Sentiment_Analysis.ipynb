```python
import pandas as pd
```

```python
df= pd.read_csv("sentiment_tweets.csv")
```

```python
list_emotion=list(df['emotion'])
list_emotion
```

```python
from collections import Counter
```

```python
count=Counter()
for emotion in df["emotion"]:
  count[emotion]+=1
```

```python
common_sen=count.most_common(20)
common_sen
```

```python
df_common_emo=pd.DataFrame(common_sen)
df_common_emo
```

```python
df_common_emo=df_common_emo.rename(columns={0:"Emotion"})
df_common_emo=df_common_emo.rename(columns={1:"count"})

df_common_emo
```

```python
import plotly.express as px
```

```python
emo=px.bar(df_common_emo, x="Emotion", y="count")
emo
```

```python
df_sentiment=df[["Text","sentiment"]]
df_sentiment.head()
```

```python
df_sentiment.info()
```

```python
positive_sen=df_sentiment[df_sentiment["sentiment"].str.contains("positive")]
```

```python
positive_sen.head()
```

```python
from collections import Counter

count = Counter()
```

```python

for text in positive_sen['Text']:
    for word in text.split(sep="@"):
        count[word] += 1
common_words=count.most_common(20)
common_words
```

```python
df.common_words=pd.DataFrame(common_words)
df.common_words.head(5)
```

```python
df.common_words=df.common_words.dropna()
```

```python
df.common_words=df.common_words.drop(0)
```

```python
df.common_words.head(3)
```

```python
df.common_words=df.common_words.rename(columns={0:"Word"})
df.common_words=df.common_words.rename(columns={1:"count"})
```
