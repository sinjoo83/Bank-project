```python
B="Bank account balance: 1"
W="Withdraw: 2"
D="Deposit: 3"
E="Exit: 4"
running = True

balance = 100


def bank_account_balance():
    print("Your balance is", balance)

def withdraw():
   global balance
   #balance = 100
   list={"1":5,"2":10,"3":20,"4":40,"5":50,"6":100}

   print("1: £5")
   print("2: £10")
   print("3: £20")
   print("4: £40")
   print("5: £50")
   print("6: £100")
   print("7: Enter value")
   sub_option=input("Select your option ")

   if int(sub_option)==7:
      value=int(input("Enter value"))
      balance = balance - value
   else:
      balance = balance - list[sub_option]


   print("The money has been withdrawn.", "Your balance is", balance)



def deposit():
    #balance = 100
    global balance
    value=int(input("Enter value"))
    balance=balance+value
    print("The money has been deposited.", "Your balance is", balance)


def exit():
    running = False
    print("exit")


pw=int(input("Enter your PIN: "))
while pw!=1111:
  pw=int(input("Enter your PIN: "))



while running:
  print(B)
  print(W)
  print(D)
  print(E)
  option=int(input("Select your option "))



  if option==1:
      bank_account_balance()
  elif option==2:
       withdraw()
  elif option==3:
       deposit()
  elif option==4:
       running = False
       print("Good bye")
  else:
        print("Invalid")





```
